```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
unrate = pd.read_csv("C:/Users/Naveen/Downloads/unrate.csv")
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.plot(unrate['DATE'].head(12),unrate['VALUE'].head(12))
```

```python
fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(5,1,1)
ax2=fig.add_subplot(5,1,2)
ax3=fig.add_subplot(5,1,3)
ax4=fig.add_subplot(5,1,4)
ax5=fig.add_subplot(5,1,5)
ax1.plot(unrate['DATE'].head(12),unrate['VALUE'].head(12))
ax2.plot(unrate['DATE'][12:24],unrate['VALUE'][12:24])
ax3.plot(unrate['DATE'][24:36],unrate['VALUE'][24:36])
ax4.plot(unrate['DATE'][36:48],unrate['VALUE'][36:48])
ax5.plot(unrate['DATE'][48:60],unrate['VALUE'][48:60])
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig=plt.figure(figsize=(10,6))
plt.plot(unrate['MONTH'][0:12],unrate['VALUE'][0:12],c='red',label='1948')
plt.plot(unrate['MONTH'][12:24],unrate['VALUE'][12:24],c='blue',label='1949')
plt.plot(unrate['MONTH'][24:36],unrate['VALUE'][24:36],c='green',label='1950')
plt.plot(unrate['MONTH'][36:48],unrate['VALUE'][36:48],c='orange',label='1951')
plt.plot(unrate['MONTH'][48:60],unrate['VALUE'][48:60],c='black',label='1952')
plt.legend(loc='upper left')
plt.show()
```

```python
reviews = pd.read_csv("C:/Users/Naveen/Downloads/fandango_Scores.csv")
reviews.head()
norm_reviews=reviews[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']]
norm_reviews.head(1)
```

```python
norm_reviews.head(10)
```

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews[num_cols].iloc[0].values
import numpy as np
bar_positions = np.arange(5) + .75
fig,ax = plt.subplots()
ax.bar(bar_positions,bar_heights,0.5)
```

```python
tick_positions = range(1,6)
fig,ax = plt.subplots()
ax.bar(bar_positions,bar_heights,0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols,rotation=90)
plt.xlabel('Rating Source')
plt.ylabel('Average Rating')
plt.title('Average User Rating For Avengers: Age of Ultron (2015)')
```
