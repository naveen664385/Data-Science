```python
import pandas as pd
import random
import numpy as np
import matplotlib as mpt
import pylab as pl
import sklearn as sk
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
data=pd.read_csv('C:/Users/bnaveen3/Desktop/SAMPLE_DATASET/diab.csv',header=None)
data[0:1][[0,1,2,3,4,5,6,7]].shape
```

```python
test_index = random.sample(range(len(data)),0)
#print(test_index)
#test_data = [data[i:1][['a','b','c','d','e','f','g','h']] for i in test_index]
train_index= [i for i in range(len(data)) if i not in test_index]
train_data = np.array([data[i:i+1][[0,1,2,3,4,5,6,7]] for i in train_index])
#train_result =[data[i:i+1][8] for i in train_index]
#print(train_result)
train_result=data[:][8]
#data[0:10][['a','b','c','d','e','f','g','h']]
#print (train_result)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=10, activation='relu', input_dim=8))
model.add(tf.keras.layers.Dense(units=1, activation='softmax'))
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
#model.compile(loss=keras.losses.binary_crossentropy,
 #             optimizer=keras.optimizers.SGD(lr=0.1, momentum=.9, nesterov=True))
```

```python
model.fit(train_data, train_result, epochs=5, batch_size=100)
```
